```python
#####################
# NORMAL RUN-3 of 4 #
#####################

myquantity = 0
myquantity2 = 0
is_my_quant_updated = True
def getquant():
    global myquantity,myquantity2
    global order_type
    global symbol_ip
    allpos = kite.positions()['net']
    for i in range(len(allpos)):
        if(allpos[i]['tradingsymbol']==symbol_ip and  allpos[i]['product']== order_type):
            myquantity = allpos[i]['quantity']
            print('this month quantity : ' + str(allpos[i]['quantity']))
            logger.debug('this_month_my_quant : '+str(myquantity))
        if(allpos[i]['tradingsymbol']==symbol_ip2 and  allpos[i]['product']== order_type):
            myquantity2 = allpos[i]['quantity']
            print('next month quantity : ' + str(allpos[i]['quantity']))
            logger.debug('next_month_my_quant : '+str(myquantity2))
getquant()

def placeneworder(quantdiff,symbol_ipp):
    global stopbuy,order_type,exchange_type,stopsell,is_my_quant_updated
    
    if(quantdiff!=0):
        is_my_quant_updated=False
    
    if(quantdiff>0):
        try:
                order_id= kite.place_order(tradingsymbol=symbol_ipp,
                            exchange=exchange_type,
                            transaction_type=kite.TRANSACTION_TYPE_BUY,
                            quantity=abs(quantdiff),
                            order_type=kite.ORDER_TYPE_MARKET,
                            variety = kite.VARIETY_REGULAR,
                            product=kite.PRODUCT_MIS if order_type=='MIS' else kite.PRODUCT_NRML)
                logger.debug('order_placed @ '+str(symbol_ipp) + ' for '+str(abs(quantdiff)))
        except Exception as e:
            print(e)
            logger.debug('order_placed_rejected @ '+str(symbol_ipp)+' for '+str(abs(quantdiff)))
    if(quantdiff<0):
        try:
                order_id= kite.place_order(tradingsymbol=symbol_ipp,
                            exchange=exchange_type,
                            transaction_type=kite.TRANSACTION_TYPE_SELL,
                            quantity=abs(quantdiff),
                            order_type=kite.ORDER_TYPE_MARKET,
                            variety = kite.VARIETY_REGULAR,
                            product=kite.PRODUCT_MIS if order_type=='MIS' else kite.PRODUCT_NRML)
                logger.debug('order_placed @ '+str(symbol_ipp) + ' for '+str(abs(quantdiff)))
        except Exception as e:
            print(e)
            logger.debug('order_placed_rejected @ '+str(symbol_ipp)+' for '+str(abs(quantdiff)))
            
    if(quantdiff!=0):
        getquant()
        is_my_quant_updated=True

def ckqnt (orderquant):
    logger.debug('in ckqnt :')
    global myquantity,myquantity2,symbol_ip,symbol_ip2,is_my_quant_updated
    quantdiff = orderquant - myquantity
    quantdiff2 = -orderquant - myquantity2
    
    logger.debug('this_month_quant_diff'+' : '+str(quantdiff) + ' = '+ str(orderquant)+'-'+str(myquantity)+' in '+str(symbol_ip))
    logger.debug('next_month_quant_diff'+' : '+str(quantdiff2) + ' = '+ str(-orderquant)+'-'+str(myquantity2)+' in '+str(symbol_ip2))
    
    y1 = threading.Thread(target=placeneworder, args=(quantdiff,symbol_ip))
    y2 = threading.Thread(target=placeneworder, args=(quantdiff2,symbol_ip2))
    
    if(is_my_quant_updated):
        y1.start()
        y2.start()
    
#     placeneworder(quantdiff,symbol_ip)
#     placeneworder(quantdiff2,symbol_ip2)
    
    
    return myquantity
```
